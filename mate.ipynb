```python
import pandas
```

```python
df = pandas.read_csv('lichess_db_puzzle.csv.zst')
```

```python
df
```

```python
df[~df['Themes'].str.contains("mate")].sort_values('Rating', ascending=False)
```

```python
compression_opts = dict(method='zip', archive_name='not-mate.csv')  
df[:int(1e6)].to_csv('not-mate-1m.zip', index=False, compression=compression_opts)  
```

```python
compression_opts = dict(method='zip', archive_name='not-mate.csv')  
df[int(1e6):2*int(1e6)].to_csv('not-mate-2m.zip', index=False, compression=compression_opts)  
```

```python
compression_opts = dict(method='zip', archive_name='not-mate.csv')  
df[2*int(1e6):].to_csv('not-mate-3m.zip', index=False, compression=compression_opts)  
```
